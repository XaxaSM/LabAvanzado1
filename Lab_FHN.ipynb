```python
# --- Gráfica 1: I vs U2 ---
plt.figure(figsize=(8,6))
plt.plot(FHNdata['U2'], FHNdata['I'], 'o--', label='Curva I vs U2',color='m' ,markersize=3)
plt.xlabel('Corriente del Colector (nA)')
plt.ylabel('Voltaje de Aceleración U2 (V)')
plt.title('Curva de Franck-Hertz función de U2')
plt.legend()
plt.grid(True)
plt.show()


```

```python
x = FHNdata['U2']
y = FHNdata['I']

coef= np.polyfit(x,y,deg=4)
poly=np.poly1d(coef)

plt.figure(figsize=(8,6))
plt.plot(x, y, 'o', markersize=3,color='m' , label='Datos')
plt.plot(x, poly(x), '-', color='deeppink', label='Ajuste Polinómico')
plt.xlabel('Corriente del Colector (nA)')
plt.ylabel('Voltaje de Aceleración U2 (V)')
plt.title('Curva de Franck-Hertz función de U2')
plt.legend()
plt.grid(True)
plt.show()

print(poly)
```

```python
FHNdata2= pd.read_excel(ruta_datos)
```

```python
# --- Gráfica 1: U1 vs U3 ---
plt.figure(figsize=(8,6))
plt.plot(FHNdata2['UB1'], FHNdata2['UA1'], 'o--', label='Datos obtenidos con el PC',color='m' ,markersize=3)
plt.xlabel('UB1')
plt.ylabel('UA1')
plt.title('Curva de Franck-Hertz')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x2 = FHNdata2['UB1']
y2 = FHNdata2['UA1']

grado = 15
coef=np.polyfit(x2,y2,grado)
poly=np.poly1d(coef)

# Graficar
plt.figure(figsize=(8,6))
plt.plot(x2, y2, 'o',color='m' , markersize=3, label='Datos obtenidos con el PC')
plt.plot(x2, poly(x2), '-', color='deeppink', label=f'Ajuste polinómico grado {grado}')
plt.xlabel('UB1')
plt.ylabel('UA1')
plt.title('Curva de Franck-Hertz (Ajuste Polinómico)')
plt.legend()
plt.grid(True)
plt.show()

#print(poly)
```
